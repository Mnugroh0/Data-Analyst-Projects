### **Customer Sales Dashboard**
> Data Preprocessing, and Cleaning. <br>
**source:** https://www.kaggle.com/datasets/onlineretailshop/online-shop-customer-sales-data 

```python
# import library
import pandas as pd
import numpy as np
import datetime
```

```python
# import dataset
data = pd.read_csv("Online Shop Customer Sales Data.csv", delimiter=",")
data.head()
# data.info()
```

```python
data.info()
```

#### **Data Cleaning and Processing**

```python
# formatting values in datasets
data["Gender"].replace([0, 1], ["Male", "Female"], inplace=True)
data["Pay_Method"].replace([0, 1, 2, 3], ["Digital Wallets", "Card", "Paypal", "Other"], inplace=True)
data["Browser"].replace([0, 1, 2, 3], ["Chrome", "Safari", "Edge", "Other"], inplace=True)
data["Newsletter"].replace([0, 1], ["Not Subscribed", "Subscribed"], inplace=True)
data["Voucher"].replace([0, 1], ["Not Used", "Used"], inplace=True)

# date
data["Purchase_DATE"] = data["Purchase_DATE"].str.replace(".", "/", regex=True)
data["Purchase_DATE"] = pd.to_datetime(data["Purchase_DATE"], format="%d/%m/%y")
data = data.loc[data["Purchase_DATE"].dt.year == 2021]
data.head()
```
